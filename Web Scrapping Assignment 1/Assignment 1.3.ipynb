## 3) Write a python program to display IMDB’s Top rated 50 Indian movies’ data (i.e. name, rating, year of release) and make data frame.


```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_movie = soup.find('td',class_="titleColumn")
first_movie
```

```python
first_movie.text
```

```python
first_movie.text.split('\n')
```

```python
first_movie.text.split('\n')[2]
```

```python
movie = []

for i in soup.find_all('td',class_="titleColumn"):
    movie.append(i.text.split('\n')[2])
movie
```

```python
yr = soup.find('span',class_="secondaryInfo")
yr.text
```

```python
year = []

for i in soup.find_all('span',class_="secondaryInfo"):
    year.append(i.text)
year
```

```python
rate = soup.find('td',class_="ratingColumn imdbRating")
rate.text
```

```python
rate.text.split('\n')
```

```python
rate.text.split('\n')[1]
```

```python
rating = []

for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.split('\n')[1])
rating
```

```python
print(len(movie),len(year),len(rating))
```

```python
import pandas as pd

df = pd.DataFrame({'Movie_Name':movie,'Ratings':rating,'Year_Of_Release':year})
df
```
