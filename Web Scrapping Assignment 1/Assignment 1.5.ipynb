## 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame

## a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
team = soup.find('tr',class_="table-body")
team.text                
```

```python
team.text.split('\n')
```

```python
team.text.split('\n')[4]
```

```python
tems = []

for i in soup.find_all('tr',class_="table-body"):
    tems.append(i.text.split('\n')[4])
tems
```

```python
mat = soup.find('tr',class_="table-body")
mat.text
```

```python
mat.text.split('\n')
```

```python
mat.text.split('\n')[7]
```

```python
matches = []

for i in soup.find_all('tr',class_="table-body"):
    matches.append(i.text.split('\n')[7])
matches
```

```python
point = soup.find('tr',class_="table-body")
point.text
```

```python
point.text.split('\n')
```

```python
point.text.split('\n')[8]
```

```python
points = []

for i in soup.find_all('tr',class_="table-body"):
    points.append(i.text.split('\n')[8])
points
```

```python
rate = soup.find('td',class_="table-body__cell u-text-right rating")
rate.text
```

```python
rating = []

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
rating
```

```python
print(len(tems),len(matches),len(points),len(rating))
```

```python
import pandas as pd

df_odi = pd.DataFrame({'Teams':tems,'Matches':matches,'Points':points,'Ratings':rating})
df_odi
```

```python
df_odi.head(10)
```

# b) Top 10 ODI Batsmen along with the records of their team andrating.

```python
from bs4 import BeautifulSoup
import requests
```

```python
pages = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
```

```python
pages
```

```python
soup = BeautifulSoup(pages.content)
soup
```

```python
first_bats = soup.find('td',class_="table-body__cell rankings-table__name name")
first_bats
```

```python
first_bats.text
```

```python
first_bats.text.split('\n')
```

```python
first_bats.text.split('\n')[1]
```

```python
batsman = []

for i in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
    batsman.append(i.text.split('\n')[1])
batsman
                   
                
```

```python
team = soup.find('span',class_="table-body__logo-text")
team.text
```

```python
teams = []

for i in soup.find_all('span',class_="table-body__logo-text"):
    teams.append(i.text)
teams
```

```python
rate = soup.find('td',class_="table-body__cell rating")
rate.text
```

```python
ratings = []

for i in soup.find_all('td',class_="table-body__cell rating"):
    ratings.append(i.text)
ratings
```

```python
print(len(batsman),len(teams),len(ratings))
```

```python
import pandas as pd

df_bats = pd.DataFrame({'Batsman':batsman,'Teams':teams,'Rating':ratings})
df_bats
```

```python
df_bats.head(10)
```
